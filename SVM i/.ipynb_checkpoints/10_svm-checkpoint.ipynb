<h1 style='color:navy' align="center">------------------------------------------</h1>

<h3 style='color:green' align="center">Lets take example of Iris flower.</h3>
<img height=200 width=300 src="iris_petal_sepal.png" />

<h3 style='color:green' align="center">Here we can easily draw multiple classification boundries to seperate  categories . But the actual problem is which one is Best for classification.</h3>



<h3 style='color:green' align="center">One way is we can select near by Data points and we can measure the distance from that data point to line. And that Distance is known as 'Margin'</h3>



<h3 style='color:green' align="center">And the line with higher Margin will be our Classification Boundry</h3>



<h3 style='color:green' align="center">And these nearby Data points are known as Support Vectors</h3>



<h3 style='color:green' align="center"> In case of 2D our boundry woulld be a line  and in case of 3D boundry would be a Plane But what if we have N-D </h3>



<h2 style='color:navy' align="center">In case of N-D that boundry is called as 'HYPER PLANE'</h2>

<h3 style='color:green' align="center">Hyper Plane is a plane in N- dimension ,thst try to seperate our different groups .</h3>

<h2 style='color:navy'>SVM draws a hyper plane in n- dimensional space such that it maximizes marzin between classification groups </h2>

```python
import seaborn as sn
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df.shape
```

```python
len(df[df.target==1])
```

```python
df[df.target==2].head(50)
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.sample(5)
```

```python
df[45:55]
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
df2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**Sepal length vs Sepal Width (Setosa vs Versicolor)**

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
```

**Train Using Support Vector Machine (SVM)**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```

**Tune parameters**

**1. Regularization (C)**
In general, regularization means to make things regular or acceptable.



<h3 style='color:green'align="center"> C means no of violations  of margin </h3>.

```python
model_C = SVC(C=7)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

**2. Gamma**

<h3 style='color:navy'align="center">In case of High Gamma we can achieve high accuracy.</h3><h4 style='color:navy'> But here we are doing too much computation and .<br> We are providing too much knowledge to model and that can cause problem of Over fitting.</h4>

```python
model_g = SVC(gamma=14)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

**3. Kernel**

<h3 style='color:navy'align="center">How we can draw a Hyperplane to this type of data distribution  ?  ?  ?  ?</h3>



<h3 style='color:navy'align="center">We should see thid from another direction.</h3>



```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```
