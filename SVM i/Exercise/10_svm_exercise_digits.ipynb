```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
#df.info()
```

```python
df['target'] = digits.target
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

**Using RBF kernel**

```python
from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
```

```python
rbf_model.fit(X_train, y_train)
```

```python
rbf_model.score(X_test,y_test)
```

**Using Linear kernel**

```python
linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)
```

```python
linear_model.score(X_test,y_test)
```

**Using 'poly' kernel**

```python
poly_model = SVC(kernel='poly')
poly_model.fit(X_train,y_train)
```

```python
poly_model.score(X_test,y_test)
```

```python
pred_rbf = rbf_model.predict(X_test)
pred_lin = linear_model.predict(X_test)
pred_poly = poly_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,pred_rbf)
cm
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(15,6))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel("True Values")
plt.ylabel("Predicted ")
```

```python
print(classification_report(y_test,pred_rbf))
```

```python
cm_lin = confusion_matrix(y_test,pred_lin)
cm_lin
```

```python
plt.figure(figsize=(15,8))
sn.heatmap(cm_lin,annot=True,fmt='d')
plt.xlabel("True Values")
plt.ylabel("Predicted ")
```

```python
print(classification_report(y_test,pred_lin))
```

```python
cm_poly = confusion_matrix(y_test,pred_poly)
cm_poly
```

```python
plt.figure(figsize=(15,8))
sn.heatmap(cm_poly,annot=True,fmt='d')
plt.xlabel("True Values")
plt.ylabel("Predicted ")
```

```python
print(classification_report(y_test,pred_poly))
```
